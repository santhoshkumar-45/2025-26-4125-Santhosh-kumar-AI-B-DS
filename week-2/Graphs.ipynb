```python
# Data Visualization using Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 6)  # X-axis: 1 to 5
y = np.array([10, 15, 7, 12, 9])  # Y-axis values
categories = ['A', 'B', 'C', 'D', 'E']

# -----------------------------
# a) Line Graph
plt.figure(figsize=(6,4))
plt.plot(x, y, marker='o', color='blue', linestyle='-')
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# b) Bar Graph
plt.figure(figsize=(6,4))
plt.bar(categories, y, color='orange', edgecolor='black')
plt.title("Bar Graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# c) Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(x, y, color='green', s=100)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#d) Histogram
data = np.random.randint(1, 20, 50)  # 50 random integers between 1 and 20
plt.figure(figsize=(6,4))
plt.hist(data, bins=8, color='purple', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# e) Pie Chart
sizes = [25, 15, 30, 10, 20]  # Percentage values
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=categories, autopct='%1.1f%%', startangle=90, colors=['red','blue','green','orange','purple'])
plt.title("Pie Chart")
plt.show()

```
