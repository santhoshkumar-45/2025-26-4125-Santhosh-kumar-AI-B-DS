```python
import matplotlib.pyplot as plt

# Line Plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("X-Y Graph")
plt.show()

# Bar Chart
subjects = ["Maths", "Physics", "Chemistry", "CS"]
marks = [80, 75, 85, 90]

plt.bar(subjects, marks, color='skyblue')
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Bar Chart")
plt.show()
```
