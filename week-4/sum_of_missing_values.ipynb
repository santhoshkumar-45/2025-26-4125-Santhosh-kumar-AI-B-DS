```python
import pandas as pd
import numpy as np

# Example DataFrame
data = {
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, np.nan],
    'C': [1, np.nan, np.nan, 4]
}

df = pd.DataFrame(data)

# Count total missing values in entire DataFrame
total_missing = df.isna().sum().sum()

print("Total missing values:", total_missing)
```
