```python
#IQR
import pandas as pd

# Example DataFrame
data = {'A': [10, 12, 14, 15, 100, 16, 18, 19, 20, 22]}
df = pd.DataFrame(data)

# Calculate Q1 and Q3
Q1 = df['A'].quantile(0.25)
Q3 = df['A'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier condition
outliers = df[(df['A'] < (Q1 - 1.5 * IQR)) | (df['A'] > (Q3 + 1.5 * IQR))]

print("Outliers detected using IQR method:")
print(outliers)
```
