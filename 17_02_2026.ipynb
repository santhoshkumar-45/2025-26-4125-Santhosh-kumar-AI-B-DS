```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data=[10,12,15,18,20,22,25,30,35,40]
df=pd.DataFrame(data,columns=["values"])
print("Dataset:")
print(df)
print("\n summary statistics: ")
print(df.describe())
print("\n variance:",np.var(data))
print("\nstandard deviation:",np.std(data))
plt.hist(data)
plt.xlabel("values")
plt.ylabel("frequency")
plt.title("Data distribution")
plt.show()
mean_value=np.mean(data)
median_value=np.median(data)
print("\n inference")
print("mean value:",mean_value)
print("median value",median_value)
if mean_value>median_value:
  print("data is slightly right_skewed")
elif mean_value<median_value:
  print("data is slightly left_skewed")
else:
  print("data is symmetric")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
url="https://raw.githubusercontent.com/dataprofessor/data/master/BostonHousing.csv"
df=pd.read_csv(url)
print("dataset(first 10 rows):")
print(df.head(10))
data=df['rm']
print("\n summary statistics:")
print(data.describe())
print("\n variance:",np.var(data))
print("\n standard deviation:",np.std(data))
plt.hist(data,bins=10,color="skyblue",edgecolor="black")
plt.axvline(np.mean(data),color="red",linestyle="dashed",linewidth=7,label=f"mean={np.mean(data):.2f}")
plt.axvline(np.median(data),color="green",linestyle="dashed",linewidth=2,label=f"meadian={np.median(data):.2f}")
plt.xlabel("rm")
plt.ylabel("frequency")
plt.title("distribution of sepal length")
plt.legend()
plt.show()
mean_value=np.mean(data)
meadian_value=np.median(data)
print("mean:",mean_value)
print("median:",meadian_value)
if mean_value>median_value:
  print("data is slightly right_skewed")
elif mean_value<median_value:
  print("data is slightly left_skewed")
else:
  print("data is symmetric")
```

```python
import pandas as pd

# data manipulation operations
data = {
    "Name": ['A', 'B', 'C'],
    "Age": [20, 21, 22],
    "Marks": [70, 80, 90]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

# adding a new column
df["Bonus"] = 5
print("\nAfter adding bonus:")
print(df)

# increase marks
df["Marks"] = df["Marks"] + 2
print("\nAfter increasing marks:")
print(df)

# create new column from existing columns
df["Total"] = df["Marks"] + df["Bonus"]
print("\nAfter adding total:")
print(df)

# change value
df.loc[0, "Age"] = 21
print("\nAfter changing Age of first student:")
print(df)

# rename column
df.rename(columns={"Total": "Final"}, inplace=True)
print("\nAfter renaming Total to Final:")
print(df)

# remove a column
df.drop("Bonus", axis=1, inplace=True)
print("\nAfter removing Bonus column:")
print(df)
```

```python
#Q3
import pandas as pd
import numpy as np
from scipy import stats

# Create sample data
data = {
    'Math': [78, 85, 96, 80, 70, 90, 88, 76, 95, 89],
    'Science': [88, 79, 92, 85, 84, 91, 87, 73, 94, 90],
    'English': [75, 82, 78, 85, 88, 90, 92, 74, 89, 84]
}

df = pd.DataFrame(data)

print("DataFrame:")
print(df)
```

```python
# Mean
mean_values = df.mean()
print("\nMean:")
print(mean_values)

# Median
median_values = df.median()
print("\nMedian:")
print(median_values)

# Mode
mode_values = df.mode()
print("\nMode:")
print(mode_values)
```

```python
#standard deviation
std_deviation = df.std()
print("\nStandard Deviation:")
print(std_deviation)
```

```python
#variance
variance = df.var()
print("\nVariance:")
print(variance)
```

```python
#kurtosis
kurtosis = df.kurt()
print("\nKurtosis:")
print(kurtosis)
```

```python
#skewness
skewness = df.skew()
print("\nSkewness:")
print(skewness)
```
