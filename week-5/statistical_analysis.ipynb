```python
import numpy as np
import math
population=np.array([10,20,30,40,50,60,70])
N=len(population)
print("\nPopulation Size(N):",N)
mu=np.mean(population)
print("Population Mean(μ):",mu)
sigma2=np.var(population,ddof=0)
print("\nPopulation Variance(σ²):",sigma2)
sigma=np.sqrt(sigma2)
print("Population Standard Deviation(σ):",sigma)
#Sample data
sample=np.array([20,30,40,50])
n=len(sample)
print("\nSample Size(n):",n)
x_bar=np.mean(sample)
print("\nSample Mean(x̄):",x_bar)
s2=np.var(sample,ddof=1)
print("\nSample Variance(s2):",s2)
s=np.sqrt(s2)
print("\nSample Standard Deviation(s):",s)

#Central Limit Theorem
#Standard Error
SE=sigma/math.sqrt(n)
print("\nStandard Error(σ/√n):",SE)

#Z-transformation
z=(x_bar-mu)/SE
print("\nZ-score:",z)
```
